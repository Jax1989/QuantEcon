```python
from __future__ import division, print_function
import numpy as np
import quantecon as qe
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
alpha = beta = 0.1
```

```python
ts_length=10000
```

```python
p = beta/(alpha+beta)
```

```python
P = [[1-alpha, alpha], [beta, 1-beta]]
```

```python
p = np.array(P)
```

```python
fig, ax = plt.subplots(figsize=(9, 6))
ax.set_ylim(-0.25, 0.25)
ax.grid()
ax.hlines(0, 0, ts_length, lw=2, alpha=0.6) #horizontal line at 0

for x0, color in ((0,'blue'),(1,'green')): 
    #generate time series data
    X = qe.mc_sample_path(P, x0, ts_length)
    X_bar = (X==0).cumsum()/(1+np.arange(ts_length))
    ax.fill_between(np.arange(ts_length), np.zeros(ts_length), X_bar - p, color=color, alpha=0.2)
    ax.plot(X_bar-p, color=color, label=r'$X_0 = \, {} $'.format(x0))
    ax.plot(X_bar-p, color='k', alpha=0.6)

ax.legend(loc='upper right')
```
