```python
import numpy as np
import matplotlib.pyplot as plt
from random import uniform
from scipy import stats
%matplotlib inline
```

```python
fig, ax = plt.subplots()
x = np.linspace(-4,4,150)
for i in range(3):
    m, s = uniform(-1,1), uniform(1,2)
    y = stats.norm.pdf(x, loc=m, scale=s)
    current_label = r'$\mu = {0:.2f}$'.format(m)
    ax.plot(x, y, lw=2, alpha=0.6, label=current_label)    
ax.legend()
plt
plt.show()
```

```python
num_col, num_row = 3, 2
fig, axes = plt.subplots(num_col, num_row, figsize=(8,12))
for i in range(num_col):
    for j in range(num_row):
        m, s = uniform(-1,1), uniform(1,2)
        x = stats.norm.rvs(loc=m, scale=s, size=100)
        axes[i,j].hist(x, bins=20, alpha=0.6)
        t = r'$\mu = {0:.1f}, \quad \sigma={1:.1f}$'.format(m,s)
        axes[i,j].set_title(t)
        axes[i,j].set_xticks([-4,0,4])
        axes[i,j].set_yticks([])
plt.show()
```
