```python
class Comsumer:
    
    def __init__(self, w):
        self.wealth = w
        
    def earn(self, y):
        self.wealth += y
        
    def spend(self, x):
        new_wealth = self.wealth - x
        if new_wealth < 0:
            print('Insufficient fund!')
        else:
            self.wealth = new_wealth
```

```python
c1 = Comsumer(10)
```

```python
c1.wealth
```

```python
c1.earn(34)
```

```python
c1.wealth
```

```python
c1.spend(1000)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class Chaos:
    def __init__(self, x0, r):
        self.x, self.r = x0, r
        
    def update(self):
        self.x = self.r * self.x * (1-self.x)
        
    def generate_sequence(self, t):
        path = []
        for i in range(t):
            path.append(self.x)
            self.update()
        return path
```

```python
ch = Chaos(0.1, 4.0)
t = 100
```

```python
fig, ax = plt.subplots()
ax.set_xlabel(r'$t$', fontsize=14)
ax.set_ylabel(r'$x_t$', fontsize=14)
x = ch.generate_sequence(t)
ax.plot(range(t), x, 'bo-', alpha=0.5, lw=2)
plt.show()
```

```python
class ECDF:
    def __init__(self, sample):
        self.observation = sample
        
    def __call__(self, x):
        sum = 0
        for i in self.observation:
            if i < x:
                sum += 1
        return sum*1.0/len(self.observation)
```

```python
from random import uniform
```

```python
samples = [uniform(0,1) for i in range(10)]
```

```python
F = ECDF(samples)
```

```python
F(0.5)
```

```python
F.observation = [uniform(0,1) for i in range(1000)]
```

```python
F(0.5)
```

```python
class Polynomial:
    def __init__(self, coeff):
        self.coeff = coeff
        
    def evaluate(self, x):
        result = 0
        for i, c in enumerate(self.coeff):
            result += c * x ** i
        return result
    
    def differentiate(self):
        new_coeff = []
        for i, c in enumerate(self.coeff):
            new_coeff.append(i*c)
        
        del new_coeff[0]
        self.coeff = new_coeff
```
